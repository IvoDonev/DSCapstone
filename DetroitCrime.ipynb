```python
import geopandas as gpd
import folium
import pandas as pd

df = pd.read_csv("data/CrimePerArea.csv")
df['_id'] = df['_id'].astype(str)
```

```python
crime_map = folium.Map([42.379858, -83.066083], zoom_start = 11)

folium.Choropleth(geo_data='data/Detroit Neighborhoods.geojson', 
                  data=df,
                  columns=['_id', 'count'],
                  key_on='feature.properties.target_fid', 
                  nan_fill_color='White',
                  fill_color='YlOrRd', 
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  highlight=True,
                  legend_name="Number of crime incidents").add_to(crime_map)

crime_map
```
