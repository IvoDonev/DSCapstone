```python
import geopandas as gpd
import folium
import pandas as pd
# Only necessary if working in a Jupyter environment.
%matplotlib inline

df = pd.read_csv("data/CrimesPerArea.csv")
df['nhood_num'] = pd.to_numeric(df["nhood_num"])
```

```python
crime_map = folium.Map([42.379858, -83.066083], zoom_start = 11)

tt = folium.Tooltip(text='hello')

folium.Choropleth(geo_data='data/Detroit Neighborhoods.geojson', data=df, columns=['nhood_name', 'numCrimes'],key_on='feature.properties.nhood_name', 
                   fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.2, highlight=True, tooltip=tt).add_to(crime_map)

crime_map
```
